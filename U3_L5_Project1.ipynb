```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

default_link = r'C:\Users\hod5\Desktop\Assignments\datasets\2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv'
# r added to beginning of the string to indicate that we want a raw value
link = default_link.replace('\\', '/')

print (link)
missing_values = ['n/a','na','--','s']
df = pd.read_csv(link,na_values= missing_values)

df.head()
```

## We will focus on Midwood High School

```python
#Create dataframe for Midwood High School
MHS = df[(df['School Name']=='Midwood High School')]
# or
# MHS = df.query("`School Name` == 'Midwood High School'")
MHS
```

```python
#Make a dataframe of Intergrated Algebra data from Midwood High School
IA = df[(df['School Name']=='Midwood High School') & (df['Regents Exam']=='Integrated Algebra')]
IA
```

```python
#Average Percent of students who scored 80% or above in Integrated Algebra by Year
IA.groupby(['Year'])[['Percent Scoring 80 or Above']].mean().sort_values('Year',ascending=False).plot(
    kind='barh',
    color='green',
    title='Average Percent of Midwood Students Who Scored 80% or Above \n in Integrated Algebra by Year (Fig 1)',
)
```

## Between 2015 and 2016, there was a double digit percentage decrease in the number of Midwood students passing Integrated Algebra with 80 or above. (Fig 1)



```python
#Total Number of Students Who Took Integrated Algebra by Year
IA.groupby(['Year'])[['Total Tested']].sum().sort_values('Year',ascending=False).plot(kind='barh', color='green',title='Total Number of Students \n Who Took Integrated Algebra by Year (Fig 2)')
```

# There is a significant decrease in the amount of students taking Integrated Algebra in 2016 in our dataset. The percentage difference shown in Fig 1 may be due to a low sample size, as shown in Fig 2

```python
# Amount of students passing Integrated Algebra
IA.groupby(['Year'])[['Percent Scoring 65 or Above']].mean().sort_values('Year',ascending=False).plot(kind='barh', color='green')
```

```python
#Percent of Midwood students who pass the Regents by Year (All subjects)
MHS.groupby(['Year'])[['Percent Scoring 65 or Above']].mean().sort_values('Year',ascending=False).plot(kind='barh', color='green')
```

```python
#Percent of Midwood students who fail the Regents by Year
MHS.groupby(['Year'])[['Percent Scoring Below 65']].mean().sort_values('Year',ascending=False).plot(kind='barh', color='green')
```

```python
# Mean Passing percentage by subject
MHS.groupby(['Regents Exam'])[['Percent Scoring 65 or Above']].mean().sort_values('Percent Scoring 65 or Above',ascending=False).plot(kind='barh', color='green', title='Mean Percent of Students That Pass \nBy Subject (Midwood High School)',legend=False)
```

## Algebra2/Trigonometry has the lowest mean passing percentage of all Regents Exams at Midwood.

```python
# Create dataframe with other High schools, exclude Midwood
otherHS = df[(df['School Level'] == 'High school') & (df['School Name'] != 'Midwood High School')]
otherHS.head()
```

```python
otherHS.groupby(['Regents Exam'])[['Percent Scoring 65 or Above']].mean().sort_values('Percent Scoring 65 or Above',ascending=False).plot(kind='barh', color='green', title='Mean Percent of Students That Pass \n By Subject (Citywide)',legend=False)
```

## Citywide, Algebra2/Trigonometry has the lowest mean passing percentage as well.

```python
MHS['Percent Scoring 65 or Above'].mean()
```

```python
otherHS['Percent Scoring 65 or Above'].mean()
```

```python
MHS2 = df.query("`School Name` == 'Midwood High School'")
MHS2['Percent Scoring 65 or Above'].mean()
```

```python
# Create new dataframes by averaging the percent of students who pass by regents subject
df1 = otherHS.groupby(['Regents Exam'])[['Percent Scoring 65 or Above']].mean()
df2 = MHS.groupby(['Regents Exam'])[['Percent Scoring 65 or Above']].mean()
```

```python
df1.rename(columns={'Percent Scoring 65 or Above':'Citywide'},inplace=True)
df1.columns
```

```python
df2.rename(columns={'Percent Scoring 65 or Above':'Midwood'},inplace=True)
df2.columns
```

```python
df3 = pd.concat([df1.transpose(), df2.transpose()]) 
df3.reset_index(inplace=True)
df3.head()
```

```python
df3.rename(columns={'index':'School'},inplace=True)
df3.head()
```

```python
sns.barplot(data = df3
            ,x = 'School'
            ,y = 'Living Environment'
           ).set_title('Mean Passing Percentage')
            
```

# Mean Living Environment Passing Percentage of Students in Midwood is higher than other City High Schools

```python
df4 = pd.concat([df1.T, df2.T]) 
df4.T
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(9, 15))

# Load the dataset
df4sorted = df4.T.sort_values("Midwood", ascending=False)

# Plot Midwood
sns.set_color_codes("muted")
sns.barplot(x="Midwood", y=df4sorted.index, data=df4sorted,
            label="Midwood High School", color="b")

# Plot OtherSchools
sns.set_color_codes("pastel")
sns.barplot(x="Citywide", y=df4sorted.index, data=df4sorted,
            label="Other High Schools", color="b").set_title('Midwood High School v Citywide \nAverage Passing Percentage (Fig3)')

# Add a legend and informative axis label
ax.legend(ncol=1, loc="lower right", frameon=True)
ax.set(xlim=(0, 100), ylabel="",
       xlabel="Average Percent of Passing Students")
sns.despine(left=True, bottom=True)
```

# Fig 3 shows Midwood High School's Average Percent of Students Passing Regents Exams are consistently above the Citywide average.
